```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path = '/content/diabetes_young_adults_india.csv'
data = pd.read_csv(file_path)
```

```python
data
```

```python
print("Dataset Info:")
print(data.info())
```

```python
numeric_data = data.select_dtypes(include=np.number)
```

```python
correlation_matrix = numeric_data.corr()
print("\nCorrelation Matrix:")
print(correlation_matrix)
```

```python
print("\nSummary Statistics:")
print(data.describe())
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data['BMI'], kde=True, bins=30, color='blue')
plt.title("Distribution of BMI")
plt.xlabel("BMI")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=data['BMI'], y=data['HbA1c'], hue=data['Gender'])
plt.title("BMI vs HbA1c")
plt.xlabel("BMI")
plt.ylabel("HbA1c")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=data['Stress_Level'], y=data['Sleep_Hours'])
plt.title("Sleep Hours by Stress Level")
plt.xlabel("Stress Level")
plt.ylabel("Sleep Hours")
plt.show()
```

```python
numeric_cols = ['Age', 'BMI', 'HbA1c', 'Fasting_Blood_Sugar', 'Cholesterol_Level']
sns.pairplot(data[numeric_cols], diag_kind='kde')
plt.show()
```
